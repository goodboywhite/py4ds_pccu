## Beautiful Soup

- 用來解析 HTML 與 XML 檔案的 Python 模組
- 先在終端機安裝模組

```bash
pip install beautifulsoup4
```

````python
import urllib.request

lalaland_url = "http://www.imdb.com/title/tt3783958/"
f = urllib.request.urlopen(lalaland_url)
html_doc = f.read()
````

````python
html_doc
````

## 運用 BeautifulSoup 函數來解析

````python
from bs4 import BeautifulSoup

soup = BeautifulSoup(html_doc, 'html.parser')
print(soup.prettify())
````

## `.select()` 與 `.text` 屬性

````python
rating_css = "strong span"
rating_list = soup.select(rating_css)
print(rating_list)
````

````python
for s in rating_list:
    print(float(s.text))
````

## 隨堂練習：從網頁中擷取出演員名單

http://www.imdb.com/title/tt3783958/

````python
def get_cast():
    cast_css = ".itemprop .itemprop"
    for s in soup.select(cast_css):
        print(s.text)
````

````python
get_cast()
````

## 隨堂練習：Yahoo! 奇摩股市

https://tw.stock.yahoo.com/d/i/rank.php?t=pri&e=tse&n=100

````python
def get_stocks():
    import urllib.request
    from bs4 import BeautifulSoup

    yahoo_stock_url = "https://tw.stock.yahoo.com/d/i/rank.php?t=pri&e=tse&n=100"
    f = urllib.request.urlopen(yahoo_stock_url)
    html_doc = f.read()
    soup = BeautifulSoup(html_doc, 'html.parser')
    stock_css = ".name"
    for s in soup.select(stock_css):
        print(s.text.split())
````

````python
get_stocks()
````

## BeautifulSoup 的缺點

- 對 CSS 選擇的支援有限
- 沒有對 XPath 選擇支援
- 必須搭配 `urllib.request.urlopen()`
- 無法解決動態網頁問題，e.g. 如果要在 IMDB 搜尋電影名稱再進行擷取
